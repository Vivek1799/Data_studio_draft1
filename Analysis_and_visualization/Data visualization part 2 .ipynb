# In this notebook , we will find out which grocery store is the most expensive and which grocery store is the least expensive

```python
import pandas as pd 
```

```python
df_final_west_eggs = pd.read_csv("egg_size_12_west.csv")
```

```python
df_final_west_eggs
```

```python
df_final_morton_eggs = pd.read_csv("egg_size_12_morton.csv")
```

```python
df_final_morton_eggs
```

```python
df_final_ctown_eggs = pd.read_csv("egg_size_12_ctown.csv")
```

```python
df_final_ctown_eggs
```

```python
df_final_foodtown_eggs = pd.read_csv("egg_size_12_foodtown.csv")
```

```python
df_final_foodtown_eggs
```

```python
df_final_west_eggs['Store'] = 'Westside'
df_final_morton_eggs['Store'] = 'Morton Williams'
df_final_ctown_eggs['Store'] = 'C-Town'
df_final_foodtown_eggs['Store'] = 'Foodtown'

df_list = [df_final_west_eggs, df_final_morton_eggs, df_final_ctown_eggs, df_final_foodtown_eggs]
df_merged = pd.concat(df_list, axis=0, ignore_index=True)

min_price = df_merged['prices'].min()
min_price_store = df_merged[df_merged['prices'] == min_price]['Store'].values[0]
min_price_product = df_merged[df_merged['prices'] == min_price]['Egg_Product'].values[0]
min_price_size = df_merged[df_merged['prices'] == min_price]['Egg_size'].values[0]

print(f"The least expensive eggs are {min_price_size} {min_price_product} from {min_price_store} with a price of {min_price}")
```

```python
df_merged
```

```python
df_merged
```

```python
df_merged.to_excel("df_merged_eggs.xlsx", index=False)
```

```python
import pandas as pd

# Group the data by store and calculate the average price of eggs
grouped = df_merged.groupby('Store').mean()

# Reset the index to make the store name a column
grouped = grouped.reset_index()

# Create a bar chart to show the average prices for each store
ax = grouped.plot.bar(x='Store', y='prices', color='red', legend=False)

# Add labels and a title to the chart
ax.set_xlabel('Store')
ax.set_ylabel('Average Price of Eggs')
ax.set_title('Average Price of Eggs by Store')
```

```python
grouped
```
