# now we will clean for Whole milk 

```python
import pandas as pd
df = pd.read_csv("whole_milk_at_food-town.csv")
```

```python
df
```

```python
df.drop(df.index[[3,6,7,15,19,20,21]], inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df['prices'] = df['prices'].str.replace('$', '').astype(float) / 100
```

```python
df
```

```python
df['Milk_size'] = df['Milk_size'].replace('0.5 gal', '64 fl oz')
```

```python
df
```

```python
df.drop(df.index[[10,16]], inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df['Milk_size'] = df['Milk_size'].replace('64 oz', '64 fl oz')
```

```python
df
```

```python
df['Milk_size'] = df['Milk_size'].replace('32 oz', '32 fl oz')
```

```python
df['Milk_size'] = df['Milk_size'].replace('32 ct', '32 fl oz')
```

```python
df
```

# now the milk is in fl oz

```python
df
```

```python
df['Milk_size'] = df['Milk_size'].str.replace('fl oz', '').astype(float)
```

```python
df_sorted = df.sort_values('prices')
```

```python
df_sorted.reset_index(drop=True, inplace=True)
```

```python
df_sorted
```

```python
df_sorted.to_csv("milk_foodtown_analysed.csv", index=False)
```
