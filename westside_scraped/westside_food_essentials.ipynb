# In this notebook, I will scrape for West side Market which is on 2840 Broadway near Columbia University for the essential food items on the instacart website.

# West side Market for eggs:

```python
from playwright.async_api import async_playwright
```

```python
playwright = await async_playwright().start()
```

```python
browser = await playwright.chromium.launch(headless = False)
```

```python
browser.new_page()
```

```python
page = await browser.new_page()
```

```python
await page.goto("https://www.instacart.com/store/westside-market/search/eggs")
```

```python
import pandas as pd

async def scrape_and_merge():
    prices = []
    egg_product = []
    egg_size = []

    for i in range(0, 24):
        try:
            price = await page.locator("div.css-1abd9id-ItemBCardDetailsSection div.css-k008qs div.css-lkyxb9-PriceIa").nth(i).inner_text()
            prices.append(price)
        except:
            continue
    df = pd.DataFrame(prices, columns=['prices'])

    for i in range(0, 24):
        try:
            name = await page.locator("h3.css-1w4764r").nth(i).inner_text()
            egg_product.append(name)
        except:
            continue
    df2 = pd.DataFrame(egg_product, columns=['Egg_Product'])

    for i in range(0, 24):
        try:
            size = await page.locator("div.css-1301wpr-Size").nth(i).inner_text()
            egg_size.append(size)
        except:
            continue
    df3 = pd.DataFrame(egg_size, columns=['Egg_size'])
    
    df4 = df.merge(df2, left_index=True, right_index=True).merge(df3, left_index=True, right_index=True)

    return df4

df4 = await scrape_and_merge()

df4
```

```python
df4.to_csv('eggs_at_westside.csv', index=False)
```
